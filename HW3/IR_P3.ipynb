# Process Data With Hazm

```python
from hazm import *
```

```python
def hazm_normalize(text):
  # preprocesing => remove '\n'
  removeNewline = text.replace("\r\n", "")

  normalizer = Normalizer()
  normalized = normalizer.normalize(removeNewline)
  normalized = normalizer.character_refinement(normalized)
  normalized = normalizer.punctuation_spacing(normalized)

  # print("(HAZM): normalized body: ", normalized)
  return normalized
```

```python
def hazm_tokenize(text):
  # to split text to words
  tokenized_words = word_tokenize(text)
  tokenized_text = sent_tokenize(text)

  # print("(HAZM): tokenized body: ", tokenized_words)
  return tokenized_words, tokenized_text
```

```python
def hazm_stem(words):
  stemmer = Stemmer()
  stemmed = []
  for word in words:
    stemmed.append(stemmer.stem(word))

  # print("(HAZM): stemmer result: ", stemmed)
  return stemmed
```

```python
def hazm_process_news(news):
  news_body = news['NewsBody']
  normalized_body = hazm_normalize(news_body)
  tokenized_body = hazm_tokenize(normalized_body)
  stemmed_body = hazm_stem(tokenized_body[0])

  return stemmed_body
```

#Clean Data

```python
for i in range(len(news)):
  if i>len(news):
    break
  if i<=len(news) and list(news[i].keys()).count('NewsBody')==0:
    del news[i]
  if str(type(news[i]['NewsBody']))=="<class 'list'>":
    news[i]['NewsBody'] =news[i]['NewsBody'][0]
```

#Making Posting-List

```python
data = {}
for i in range(len(news)):
  docId = i+1
  tokens = hazm_process_news(news[i])
  tokens_freq = {}
  for token in tokens:
    if len(token)>=3:
      if token in tokens_freq:
        tokens_freq[token] += 1
      else:
        tokens_freq[token]= 1
  for token in tokens_freq:
    if token in data:
      data[token][0][1] += tokens_freq[token]
      data[token][1][1].append([docId, tokens_freq[token]])
    else:
      data[token] = [['freq', tokens_freq[token]],
                     ['docs',  [[docId, tokens_freq[token]] ]]
                     ]
print(len(data))
print(data)
```

#Calculating the output of each query
Return a list of docID

```python
def operation(arr):
  output = np.ones(1000);
  for item in arr:
    # print(item)
    if item.find("!")==0:
      item = item.replace("!","")
      if item in df.columns:
        output = output & np.logical_not(df[item])
      else:
        print(f'No "{item}" as columns')
        return;
    else:
      if item in df.columns:
        output = output & df[item]
      else:
        print(f'No "{item}" as columns')
        return;

  print([i for i in range(len(output)) if output.iloc[i]])
```
